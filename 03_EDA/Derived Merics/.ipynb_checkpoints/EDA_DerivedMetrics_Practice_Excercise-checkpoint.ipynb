```python
df1 = pd.read_csv('odi-batting.csv')
```

```python
df1['Centuries'] = df1['Runs'].apply(lambda x : 1 if ( x >= 100) else 0)
```

```python
df1.head()
```

```python
df1 = (df1.groupby(by=df1['Player'])).sum()
```

```python
df1.sort_values(by=['Centuries'],ascending=False, inplace=True)
```

```python
df1
```

```python
df2 = pd.read_csv('odi-batting.csv')
```

```python
df2['Centuries'] = df2['Runs'].apply(lambda x : 1 if ( x >= 100) else 0)
```

```python
df2 = df2[df2['Centuries'] == 1]
```

```python
df2['Strike Rate'] = df2['Runs']/df2['Balls']
```

```python
df2.sort_values(by=['Strike Rate'],ascending=False, inplace=True)
```

```python
df2.head(100)
```

```python
df3 = pd.read_csv('odi-batting.csv')
```

```python
df3['Centuries'] = df3['Runs'].apply(lambda x : 1 if ( x >= 100) else 0)
```

```python
df3 = df3[df3['Country'] == 'India']
```

```python
from datetime import datetime
```

```python
df3['Year'] = pd.DatetimeIndex(df3['MatchDate']).year
```

```python
df3.head(10)
```

```python
df3 = (df3.groupby(by=df3['Year'])).sum()
```

```python
df3.head(100)
```

```python
df3.sort_values(by=['Centuries'],ascending=False, inplace=True)
```
