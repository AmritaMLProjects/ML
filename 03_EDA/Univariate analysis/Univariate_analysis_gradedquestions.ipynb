```python
import pandas as pd
```

```python
df = pd.read_csv('EDA_census.csv')
```

```python
df
```

```python
df_f = df[(df["Age-group"]=='20-24') & (df["Area Name"] == 'INDIA') & (df["Total/Rural/urban"]=='Total')]
```

```python
df_f
```

```python
(df_f['Total Illiterate Females']/df_f['Total Females'])*100
```

```python
df_byagegroups = df[(df["Area Name"] == 'INDIA') & (df["Total/Rural/urban"]=='Total')]
```

```python
df_byagegroups['litrecyrate'] = (df_byagegroups['Total literate Persons']/df_byagegroups['Total Persons'])*100
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_byagegroups['Age-group'], df_byagegroups['litrecyrate'], alpha= 0.7, s = 50 )
plt.xticks(rotation=90)
plt.show()
```
