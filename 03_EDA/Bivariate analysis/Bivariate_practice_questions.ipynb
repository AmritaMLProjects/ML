```python
import pandas as pd
```

```python
cur = pd.read_csv('currencies.csv')
```

```python
cur.head()
```

```python
cur.corr()
```

```python
import matplotlib.pyplot as plt

f = plt.figure(figsize=(40, 40))
plt.matshow(cur.corr(), fignum=f.number)
plt.xticks(range(cur.select_dtypes(['number']).shape[1]), cur.select_dtypes(['number']).columns, fontsize=20, rotation=90)
plt.yticks(range(cur.select_dtypes(['number']).shape[1]), cur.select_dtypes(['number']).columns, fontsize=20, rotation=0)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=20)
plt.title('Correlation Matrix', fontsize=20);
```
