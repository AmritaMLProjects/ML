# Data Preprocessing

```python
#Keep numeric features
train = train.select_dtypes(['int64', 'float64'])

#Drop features that are ids
train = train.drop(columns = ['id','circle_id'])

#Drop features with missing values
cols_with_missing = [col for col in train.columns if train[col].isnull().any()]
train = train.drop(cols_with_missing, axis=1)

train
```

# Predictions

The first model will be a decision tree.  I will tune the max number of leaf nodes, and then use a random forest regressor as the final model.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from sklearn import tree

train_y = train['churn_probability']
train_X = train.drop(columns = ['churn_probability'])

train_X, val_X, train_y, val_y = train_test_split(train_X, train_y, random_state = 1)

clf = DecisionTreeClassifier(random_state=1, max_leaf_nodes=3)
model = clf.fit(train_X, train_y)
test_predictions = clf.predict(val_X)
print("Accuracy Score: ", accuracy_score(val_y, test_predictions))
```

Plotting the decision tree shows that customers with no minutes of usage for incoming calls in August, and no 2G mobile internet usage in August, are predicted by this model to leave.

```python
from matplotlib import pyplot as plt

feature_names=train_X.columns.tolist()

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, feature_names=feature_names, class_names=['Left','Stayed'], filled=True)
```

The accuracy can be improved a little by increasing the number of max leaf nodes to 10.

```python
for max_leaf_nodes in range(2, 31, 2):
    clf = DecisionTreeClassifier(random_state=1, max_leaf_nodes=max_leaf_nodes)
    model = clf.fit(train_X, train_y)
    predictions = clf.predict(val_X)
    print(f"Accuracy score for {max_leaf_nodes} max leaf nodes: ", accuracy_score(val_y, predictions))
```

The random forest regressor does not increase in accuracy, but will be used as the final model.

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1, max_leaf_nodes=10)
forest_model.fit(train_X, train_y)
predictions = np.around(forest_model.predict(val_X))
print("Accuracy Score: ", accuracy_score(val_y, predictions))
```

```python
test = pd.read_csv('/kaggle/input/telecom-churn-case-study-hackathon-c51/test.csv')
test_features = test[feature_names]
test_predictions = np.around(clf.predict(test_features))
test['churn_probability'] = test_predictions
submission = test[['id','churn_probability']]
submission.to_csv('submission.csv', index=False)
```
