Data Augmentation with Keras

https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator


```python
%matplotlib inline

import os
import numpy as np
import tensorflow as tf

from PIL import Image
from matplotlib import pyplot as plt
```

```python
print('Using TensorFlow', tf.__version__)
```

```python

image_path = 'C:/Users/xxxx/Desktop/xxxx/D/train/cat/Cat.jpg'

plt.imshow(plt.imread(image_path));
```

#### Rotation - 50

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(rotation_range = 50) 
x, y = next(generator.flow_from_directory('C:/Users/xxxx/Desktop/xxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

#### Brightness

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
                                        brightness_range = (0.5, 5))

x, y = next(generator.flow_from_directory('C:/Users/xxxx/Desktop/xxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

#### Shear Transformation

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
                                        shear_range=40)

x, y = next(generator.flow_from_directory('C:/Users/xxxxx/Desktop/xxxxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

#### Zoom

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
                                        zoom_range = 2)

x, y = next(generator.flow_from_directory('C:/Users/xxxxx/Desktop/xxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

#### Flips and rotate

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
                                    horizontal_flip = True,
                                    vertical_flip = True,
                                    rotation_range = 30)

x, y = next(generator.flow_from_directory('C:/Users/xxxx/Desktop/xxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

#### Width and Height Shifts

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
                                    width_shift_range = [-100, -50, 0, 50, 100],
                                    height_shift_range = [-50, 0, 50])

x, y = next(generator.flow_from_directory('C:/Users/xxxx/Desktop/xxxx/D', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```
